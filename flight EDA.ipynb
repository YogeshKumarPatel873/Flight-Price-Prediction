```python
def minutee(time):
    x=time.split(" ")
    if len(x)==2:
        hour,minut=time.split(" ")
        hour=int(hour.split('h')[0])
        minut=int(minut.split('m')[0])
        total_time=(hour*60)+minut
        return total_time
    elif 'h' in time:
        hour=time.split('h')[0]
        hour=int(hour)
        total_time=hour*60
        return total_time
    else:
        minut=time.split('m')[0]
        total_time=int(minut)
        return total_time
```

```python
final_df["Duration"]=final_df["Duration"].map(minutee)
```
